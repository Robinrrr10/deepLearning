### Tutorial 8 : Activation functions

##### This has python code for some activation functions. Just for our understanding.

```python
import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))
    # This gives number between 0 to 1
```

```python
sigmoid(10)
```

```python
sigmoid(78)
```

```python
sigmoid(-8)
```

```python
sigmoid(-50)
```

```python
def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
    # This gives number between -1 to 1
```

```python
tanh(40)
```

```python
tanh(7)
```

```python
tanh(-5)
```

```python
tanh(-40)
```

```python
def relu(x):
    return max(0, x)
    # This gives 0 or the number itself
```

```python
relu(6)
```

```python
relu(39)
```

```python
relu(-4)
```

```python
relu(-67)
```

```python
def leaky_relu(x):
    return max(0.1 * x, x)
    # This gives the number * 0.1 or the number itself
```

```python
leaky_relu(5)
```

```python
leaky_relu(81)
```

```python
leaky_relu(-5)
```

```python
leaky_relu(-40)
```

```python
leaky_relu(-74)
```
