### Tutorial 11 : Loss and Cost functions

```python
import numpy as np
```

```python
y_predicted = np.array([1, 1, 0, 0, 1])
y_true = np.array([0.3, 0.7, 1, 0, 0.5])
```

```python
def mean_absolute_error(y_true, y_predicted):         # Function which explain how the mean absoulte error works
    total_errors = 0
    for yt, yp in zip(y_true, y_predicted):
        error = abs(yt - yp)
        print(error)
        total_errors = total_errors + error
    print("Total erros:" + str(total_errors))
    mean_absolute_error_value = total_errors / len(y_true)
    print("MAE:" + str(mean_absolute_error_value))
    return mean_absolute_error_value
```

```python
mean_absolute_error(y_true, y_predicted)
```

```python
def log_loss(y_true, y_predicted):
    y_predicted_new = [max(i, epsilon) for i in y_predicted]
    y_predicted_new = [min(i, 1 - epsilon) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    return -np.mean(y_true * np.log(y_predicted_new) + (1 - y_true) * np.log(1 - y_predicted_new))
```

```python
log_loss(y_true, y_predicted)
```
