```python
import numpy as np
```

```python
y_predicted = np.array([1, 1, 0, 0, 1])
y_true = np.array([0.3, 0.7, 1, 0, 0.5])
```

```python
def mean_squared_error(y_true, y_predicted):         # Function which explain how the mean squared error works
    total_errors = 0
    for yt, yp in zip(y_true, y_predicted):
        #error = ( yt * yt ) + (yp * yp) - 2 * yt * yp    # This and below are same. Here we have give the formula of a - b whole squre. Below we directly applied. thats it
        error = (yt - yp)**2
        print(error)
        total_errors = total_errors + error
    print("Total erros:" + str(total_errors))
    mean_squared_error_value = total_errors / len(y_true)
    print("MSE:" + str(mean_squared_error_value))
    return mean_squared_error_value
```

```python
mean_squared_error(y_true, y_predicted)
```
