```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
X_train_scalled = X_train/255
X_test_scalled = X_test/255
```

```python
X_train_scalled
```

```python
X_train_scalled[0]
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
tb_callback = tf.keras.callbacks.TensorBoard(log_dir='logs/adam', histogram_freq=1)
```

```python
model.fit(X_train_scalled, y_train, epochs=5, callbacks=[tb_callback])
```

```python
model.evaluate(X_test_scalled, y_test)
```

```python
model2 = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])
```

```python
model2.compile(optimizer='SGD', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
tb_callback = tf.keras.callbacks.TensorBoard(log_dir='logs/sgd', histogram_freq=1)
```

```python
model2.fit(X_train_scalled, y_train, epochs=5, callbacks=[tb_callback])
```

```python
model2.evaluate(X_test_scalled, y_test)
```
