### Tutorial 10: Matrix

```python
import numpy as np
```

```python
revenue = np.array([[100, 120, 135], [40, 32, 34], [42, 55, 60]])
expense = np.array([[90, 96, 111], [20, 15, 23], [32, 40, 41]])
```

```python
profit_and_loss = revenue - expense        # This is the power of numpy, we substract/add like this. it will do for matrix
profit_and_loss
```

```python
price = np.array([1000, 400, 1500])
units = np.array([[80, 94, 105], [40, 25, 15], [50, 60, 88]])
```

```python
units_price = price * units
```

```python
units_price
```

#### IT tool the same row and multipled. Above this method is called broadcasting. Meaning it is trying to match with bigger size array.
##### Here is it multiplying each with same price
##### When space of matrix is not same, it will expand and match with bigger. This is broadcasting. By default it will do this
##### In above even if you give same price with 2 more rows, it will give same result

### But we have to get the price for each year. Where we need to match with smaller size. Below is the code

```python
price_of_units_in_each_years = np.dot(price, units)  # WE can use this dot function to do that
price_of_units_in_each_years
```

##### numpy has this dot method for using to calculate dot product
